```python
import pandas as pd
import socket
import time

# Load CSV
df = pd.read_csv("group05_http_input.csv")
df
```

```python
def encapsulate(msg):
    return {
        "Application": msg,
        "Transport": "TCP Header",
        "Network": "IP Header",
        "Link": "Ethernet Header"
    }

for _, row in df.iterrows():
    encapsulated = encapsulate(row['message'])
    print(encapsulated)
```

```python
HOST = "127.0.0.1"
PORT = 8080

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

try:
    s.connect((HOST, PORT))
    print("Connected.")

    for _, row in df.iterrows():
        msg = row['message']
        s.send(msg.encode())
        time.sleep(0.2)

    s.close()

except Exception as e:
    print("Error:", e)
```
